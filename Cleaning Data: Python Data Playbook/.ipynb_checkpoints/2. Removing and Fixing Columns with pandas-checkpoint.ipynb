## 2.2 Changing Columns Casing

```python
import pandas as pd
```

```python
data = pd.read_csv('_datasets/artwork_sample.csv')
```

```python
data.columns
```

```python
data.columns.str.lower()
```

```python
[x.lower() for x in data.columns]
```

```python
data.columns
```

```python
data.columns = [x.lower() for x in data.columns]
```

```python
data.columns
```

```python
data = pd.read_csv('_datasets/artwork_sample.csv')
```

```python
data.columns
```

```python
import re
```

```python
data.columns = [re.sub(r'([A-Z])', r'_\1', x).lower() for x in data.columns]
```

```python
data.columns
```

## 2.3 Renaming Columns

```python
import pandas as pd
```

```python
data = pd.read_csv('_datasets/artwork_sample.csv')
```

```python
data.columns
```

```python
data.rename(columns={'thumbnailUrl': 'thumbnail'})
```

```python
data.rename({'thumbnailUrl': 'thumbnail'}, axis=1)
```

```python
data
```

```python
data.rename(columns={'thumbnailUrl': 'thumbnail'}, inplace=True)
```

```python
data
```

```python
data.rename(columns=lambda x: x.lower(), inplace=True)
```

```python
data.columns
```
