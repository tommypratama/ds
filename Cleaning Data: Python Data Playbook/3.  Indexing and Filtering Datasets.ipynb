## 3.4 Filtering Data with str.contains

```python
import pandas as pd
```

```python
data = pd.read_csv('_datasets/artwork_sample.csv')
```

```python
data.head()
```

```python
data.medium
```

```python
type(data.medium)
```

```python
data.medium.str
```

```python
data.medium.str.contains('Graphite')
```

```python
data.loc[data.medium.str.contains('Graphite'), ['artist', 'medium']]
```

```python
data.head()
```

```python
data.loc[data.medium.str.contains('Graphite', case=False), ['artist', 'medium']]
```

```python
data.loc[data.medium.str.contains('(?i)Graphite', regex=True), ['artist', 'medium']]
```

```python
data.loc[data.medium.str.contains('(?i)Graphite', regex=True) | data.medium.str.contains('(?i)Line', regex=True), ['artist', 'medium']]
```

```python
data.loc[data.medium.str.contains('graphite|line', case=False, regex=True), ['artist', 'medium']]
```

```python
data.dtypes
```

```python
data.year.str.contains('1826')
```

```python
data.year.astype(str).str.contains('1826')
```

```python
data.loc[data.year.astype(str).str.contains('1826')]
```
