## 1.3 Normalizing Data

```python
import pandas as pd
```

```python
data = pd.read_csv('artwork_sample.csv')
```

```python
data.head()
```

```python
data.dtypes
```

```python
data.height.mean()
```

```python
data.artist.mean()
```

```python
data.height
```

```python
data.height.min()
```

```python
data.height.max()
```

```python
data.height.mean()
```

```python
data.height.std()
```

```python
height = data.height
```

```python
norm = (height - height.mean()) / height.std()
norm
```

```python
minmax = (height - height.min()) / (height.max() - height.min())
minmax
```

```python
minmax.min()
```

```python
minmax.max()
```

```python
data
```

```python
data.height = minmax
```

```python
data['standardized_height'] = norm
```

```python
data.new_column = norm
```

```python
data.head()
```
